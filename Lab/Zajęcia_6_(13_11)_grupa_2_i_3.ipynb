# Dodatkowe informacje o s≈Çownikach


1. Zbuduj s≈Çownik z ciƒÖgu znak√≥w.
Masz string: "a=1;b=2;c=3".
Zamie≈Ñ go na s≈Çownik: {'a': 1, 'b': 2, 'c': 3}.

```python
# zmienianie ciƒÖgu znak√≥w na s≈Çownik w wersji rozwlek≈Çej

d = {}
s = "a=1;b=2;c=3"
ls = s.split(";")
print(ls)
new_ls = []
for x in ls:
  new_ls.append(x.split("="))

print(new_ls)

for l in new_ls:
  d[l[0]] = int(l[1])

d
```

```python
# zmienianie ciƒÖgu znak√≥w na s≈Çownik w wersji "for comprehension"


s = "a=1;b=2;c=3"
result = {k:int(v) for k, v in (item.split('=') for item in s.split(';'))}
result
```

## Definicja silni
n! = n √ó (n‚àí1)!, dla ùëõ > 1 n>1, a 0!= 1

```python
def silnia_rec(n):
    print(f"Wyo≈Çanie silnia_rec({n})")
    if n == 0:
        print(f"Koniec wywo≈Çania silnia({n}), zwracana warto≈õƒá: {1}")
        return 1
    x =  n*silnia_rec(n-1)
    print(f"Koniec wywo≈Çania silnia({n}), zwracana warto≈õƒá: {x}")
    return x
```

```python
silnia_rec(5)
```

## Rekurencyjna funkcja sumy

Wykorzystaj schemat: head and tail: head = pierwszy element listy, tail: lista bez pierwszego elementu.

```python
def sum_rec(l: list) -> int:
  if l == []:
    return 0
  return l[0] + sum_rec(l[1:])


sum_rec([1,2,3,4,5])
```

## CiƒÖg Fibonacciego

Ka≈ºdy wyraz to suma dw√≥ch poprzednich:

ùêπ(ùëõ)=ùêπ(ùëõ‚àí1)+ùêπ(ùëõ‚àí2)F(n)=F(n‚àí1)+F(n‚àí2), z ùêπ(0)=0, F(1)=1

Fibonacci: 1,1,2,3,5,8,13,21...

```python
def fib(i: int) -> int:
  # if i in [1,2]
  #   return 1

  # if i <= 1
  #   return i

  if i == 1 or i == 2:
    return 1

  return fib(i-1) + fib(i-2)

fib(5)
```

```python
# optymalizacja rekurencji za pomocƒÖ lru_cache

from functools import lru_cache

@lru_cache(None)
def fib(i: int) -> int:
  print(f"wywo≈Çanie fib {i}")
  if i <= 1:
    print(f"koniec wywo≈Çania fib {i}, zwraca 1")
    return i
  x = fib(i-1) + fib(i-2)
  print(f"koniec fib {i}, zwraca warto≈õƒá {x}")
  return x


fib(5)
```

```python
# memoizacja - optymalizacja rekurencji z pomocƒÖ wbudowanego s≈Çownika zapamiƒôtujƒÖcego obliczone warto≈õci

def fib_dict(i:int, d: dict = {0:0, 1:1}) -> int:
  if i not in d:
    print(f"stan s≈Çownika przy {i}, to {d}")
    d[i] = fib_dict(i - 1) + fib_dict(i -2)
    print(f"stan s≈Çownika przy {i}, to {d}")
  return d[i]

fib_dict(5)
```

```python
# memoizacja za pomocƒÖ s≈Çownika, k√≥ry jest inicjowany od nowa przy ka≈ºdym wywo≈Çaniu funkcji (nie przechwuje w pamiƒôci niepotrzebnych warto≈õci )

def fib_dict_none(i:int, d: dict = None) -> int:

  if d == None:
    d = {0:0, 1:1}


  if i not in d:
    print(f"stan s≈Çownika przy {i}, to {d}")
    d[i] = fib_dict_none(i - 1) + fib_dict_none(i -2)
    print(f"stan s≈Çownika przy {i}, to {d}")
  return d[i]

fib_dict_none(5)
```

LRU (Least Recently Used) zapamiƒôtuje wyniki funkcji.

Funkcja nie musi liczyƒá tego samego wiele razy

Wynik zostaje w pamiƒôci

Optymalizuje czas oblicze≈Ñ.

```python
# Bƒôdziemy u≈ºywaƒá dekoratora funkcji, jest to funkcja, kt√≥ra modyfikuje inne funkcje

def dekorator(f):
    def wrapper():
        print("Zaczynam‚Ä¶")
        wynik = f()
        print("Koniec.")
        return wynik
    return wrapper

@dekorator
def hello():
    print("Cze≈õƒá!")

hello()
```

```python
import time
from functools import lru_cache

# --- Funkcja rekurencyjna bez cache ---
def fib_no_cache(n):
    if n < 2:
        return n
    return fib_no_cache(n-1) + fib_no_cache(n-2)

# --- Funkcja rekurencyjna z cache ---
@lru_cache(maxsize=None)
def fib_with_cache(n):
    if n < 2:
        return n
    return fib_with_cache(n-1) + fib_with_cache(n-2)

# ----Moja funkja ze s≈Çownikiem

def fib_dict(i:int, d: dict = {0:0, 1:1}) -> int:
  if i not in d:
    d[i] = fib_dict(i - 1) + fib_dict(i -2)
  return d[i]

# --- Pomiar czasu ---
def measure_time(func, *args):
    start = time.perf_counter()
    result = func(*args)
    end = time.perf_counter()
    print(f"{func.__name__}({args[0]}) = {result}")
    print(f"Czas wykonania: {end - start:.6f} sekundy\n")



measure_time(fib_no_cache, 35)
measure_time(fib_with_cache, 35)
measure_time(fib_dict, 35)
```
