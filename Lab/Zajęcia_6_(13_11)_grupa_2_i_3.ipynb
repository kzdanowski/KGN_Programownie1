# Dodatkowe informacje o sÅ‚ownikach


1. Zbuduj sÅ‚ownik z ciÄ…gu znakÃ³w.
Masz string: "a=1;b=2;c=3".
ZamieÅ„ go na sÅ‚ownik: {'a': 1, 'b': 2, 'c': 3}.

```python
# zmienianie ciÄ…gu znakÃ³w na sÅ‚ownik w wersji rozwlekÅ‚ej

d = {}
s = "a=1;b=2;c=3"
ls = s.split(";")
ls
new_ls = []
for x in ls:
  y = x.split("=")
  new_ls.append(y)

new_ls

for l in new_ls:
  d[l[0]] = int(l[1])

d
```

```python
# zmienianie ciÄ…gu znakÃ³w na sÅ‚ownik w wersji "for comprehension"


s = "a=1;b=2;c=3"
result = {k:int(v) for k, v in (item.split('=') for item in s.split(';'))}
result
```

```python
# kkonstrukcja sÅ‚ownika z listy elelemntÃ³w z pomocÄ… funkcji "setdefault"

dane = [("Warszawa", "Anna"), ("KrakÃ³w", "Piotr"), ("Warszawa", "Ola")]
d = {}
for k, v in dane:
    d.setdefault(k, []).append(v)

d
```

LRU (Least Recently Used) zapamiÄ™tuje wyniki funkcji.

Funkcja nie musi liczyÄ‡ tego samego wiele razy

Wynik zostaje w pamiÄ™ci

Optymalizuje czas obliczeÅ„.

```python
import time
from functools import lru_cache

# --- Funkcja rekurencyjna bez cache ---
def fib_no_cache(n):
    if n < 2:
        return n
    return fib_no_cache(n-1) + fib_no_cache(n-2)

# --- Funkcja rekurencyjna z cache ---
@lru_cache(maxsize=None)
def fib_with_cache(n):
    if n < 2:
        return n
    return fib_with_cache(n-1) + fib_with_cache(n-2)

# --- Pomiar czasu ---
def measure_time(func, *args):
    start = time.perf_counter()
    result = func(*args)
    end = time.perf_counter()
    print(f"{func.__name__}({args[0]}) = {result}")
    print(f"Czas wykonania: {end - start:.6f} sekundy\n")



measure_time(fib_no_cache, 35)
measure_time(fib_with_cache, 35)
```
