# Kontynuacja przetwarzania plików z wykorzystaniem wyjątków

```python
# skrypt do stworzenia pliku tekstowego

with open("liczby.txt", "w") as f:
  f.write("""1
23
33
5
6
7""")
```

```python
# funkcja licząca lizbę wierszy, znaków i słów w pliku, z obsługą wyjątków

def file_stats(path):
  try:
    with open(path, "r") as f:
      lines = f.readlines()
      num_lines = len(lines)
      num_words = 0
      for line in lines:
        num_words += len(line.split())
      with open(path) as f:
        all = f.read()
      num_chars = len(all)
    print(f"Liczba linii: {num_lines}")
    print(f"Liczba znaków: {num_chars}")
    print(f"Liczba słów: {num_words}")
    return num_lines, num_chars, num_words
  except FileNotFoundError:
    print(f"Plik {path} nie istnieje")
    return None
  except PermissionError:
    print(f"Brak uprawnień do pliku {path}")
    return None
  except Exception as e:
    print(f"Wystąpił błąd: {e}")
    return None
```

```python
file_stats("liczby.txt")
```
