```python
import pandas as pd

# 1. Tworzymy dane
df = pd.DataFrame({
    'produkt': ['A', 'B', 'C'],
    'cena': [50, 150, 200],
    'ilosc': [10, 2, 1]
})

# 2. Filtrujemy drogie produkty
drogie = df[df['cena'] > 100]

# 3. Liczymy całkowity przychód z tych produktów
suma = (drogie['cena'] * drogie['ilosc']).sum()
print(suma)
```
