```python
def znajdz_pozycje_slowa(sciezka_pliku, szukane_slowo):
    """
    Otwiera plik i szuka pierwszego wystƒÖpienia podanego s≈Çowa.
    Wypisuje numer linii (liczƒÖc od 1) i indeks znaku w tej linii.
    """
    try:
        # Otwieramy plik w trybie odczytu ('r') z kodowaniem utf-8
        with open(sciezka_pliku, 'r') as plik:

            # enumerate pozwala nam iterowaƒá po liniach i jednocze≈õnie mieƒá ich numer

            for numer_linii, tre≈õƒá_linii in enumerate(plik):

                # Metoda .find() zwraca indeks pierwszego wystƒÖpienia lub -1, je≈õli nie znajdzie
                indeks = tre≈õƒá_linii.find(szukane_slowo)

                if indeks != -1:
                    print(f"Znaleziono s≈Çowo '{szukane_slowo}': Linia {numer_linii}, Indeks {indeks}")
                    return (numer_linii, indeks) # Zwracamy krotkƒô z wynikiem i ko≈Ñczymy funkcjƒô

        # Je≈õli pƒôtla siƒô sko≈Ñczy i nic nie zwr√≥ci≈Ça, to s≈Çowa nie ma
        print(f"S≈Çowo '{szukane_slowo}' nie wystƒôpuje w tym pliku.")
        return None

    except FileNotFoundError:
        print(f"B≈ÇƒÖd: Nie znaleziono pliku pod ≈õcie≈ºkƒÖ: {sciezka_pliku}")
    except Exception as e:
        print(f"WystƒÖpi≈Ç nieoczekiwany b≈ÇƒÖd: {e}")
```

```python
def safe_open(file, n):
  try:
    f = open(file, n)
    return f
  except FileNotFoundError:
    print(f"B≈ÇƒÖd: Nie znaleziono pliku")
    return None
  except PermissionError:
    print(f"B≈ÇƒÖd: Nie masz uprawnie≈Ñ do pliku")
    return None
  except EncodingWarning:
    print(f"B≈ÇƒÖd: Nieprawid≈Çowy kodowanie pliku")
    return None
```

```python
def del_empty_lines(plik_in, plik_out):
  with open(plik_in, "r") as source:
    with open(plik_out, "w") as end:
      for line in source:
        if line.strip() != "":
          end.write(line)

```

```python
import os

def test_zadanie():
    katalog = "dane"
    plik_stary = os.path.join(katalog, "informacje.txt")
    plik_nowy = os.path.join(katalog, "info.txt")

    print("üîç Sprawdzanie zadania...\n")

    # 1. Sprawdzenie katalogu
    if not os.path.isdir(katalog):
        print("‚ùå Brak katalogu 'dane'.")
        return
    print("‚úî Katalog 'dane' istnieje.")

    # 2. Sprawdzenie pliku info.txt (po zmianie nazwy)
    if not os.path.isfile(plik_nowy):
        if os.path.isfile(plik_stary):
            print("‚ùå Znaleziono 'informacje.txt', ale nie zmieniono jego nazwy na 'info.txt'.")
        else:
            print("‚ùå Brak pliku 'info.txt'.")
        return
    print("‚úî Plik 'info.txt' istnieje.")

    # 3. Sprawdzenie zawarto≈õci pliku
    with open(plik_nowy, "r", encoding="utf-8") as f:
        lines = f.read().strip().splitlines()

    if len(lines) < 4:
        print("‚ùå Plik nie zawiera wszystkich wymaganych linii (3 linie + linia dopisana przez program).")
        return
    print("‚úî Plik zawiera wystarczajƒÖcƒÖ liczbƒô linii.")

    # 4. Sprawdzenie dopisanej linii
    if lines[-1] != "To jest nowa linia dopisana przez program.":
        print("‚ùå Ostatnia linia nie zosta≈Ça dopisana poprawnie.")
        return
    print("‚úî Dopisana linia jest poprawna.")

    # 5. Wy≈õwietlenie listy plik√≥w (opcjonalne sprawdzenie)
    pliki = os.listdir(katalog)
    if len(pliki) == 0:
        print("‚ùå Katalog 'dane' jest pusty ‚Äî co≈õ jest nie tak.")
        return

    print("‚úî W katalogu znajduje siƒô co najmniej jeden plik.")
    print("\nüéâ Wszystkie testy zaliczone poprawnie!")

# Uruchom test
if __name__ == "__main__":
    test_zadanie()
```
